```python
#Open the csv file -- budget_data.csv (composed of two columns, Date and Profit/Losses).
#Import numpy for mathematical function needs
import csv
import numpy

#set up variables needed for final outputs (months, p&l, profit, loss, lists and max/min variables)
total_months = 0
total_p_and_l = 0
list_of_dict = []
list_of_change = []
maximum_change = 0
minimum_change = 0
maximum_change_date = ""
minimum_change_date = ""

#OPEN FILE
#-------------------------------------------------------
#open file by reading, using WITH statement to avoid needing to close and as a dictionary.  Name file as "File" for easy reference
with open('Resources/budget_data.csv', newline='') as file:
    reader = csv.DictReader(file)
    #Create function to create new list of dictionaries    
    def dictmake(date,p_and_l):
        date = date
        p_and_l = p_and_l
        #Create the new dictionary based off inputs
        dict_of_pl = {
            "date" : date,
            "amount" : p_and_l,
       }
        #Append dictionary to a new list of dictionaries
        list_of_dict.append(dict_of_pl)

#CALCULATE TOTAL ENTRIES, P&L TOTAL AND CREATE NEW LIST OF CHANGES
#-------------------------------------------------------
    #Look at each row in the CSV file and calculate needed information for number of months and P&L total.  Change to integers where needed
    for row in reader:
        #check for accuracy using, hidden based on assignment: print(row['Date'],row['Profit/Losses'])
        total_months = total_months+1
        total_p_and_l = total_p_and_l + int(row['Profit/Losses'])
        #Use custom function to create new dictionary entries into new list
        dictmake(row['Date'],row['Profit/Losses'])
    #Go through new list of dictionaries to calculate change between numbers and assign to date of new month, assign numbers to new list of dictionaries before calculating average
    #Use i as a loop in the length or number of dictionaries
    for i in range(1, len(list_of_dict)):
        #generate new variables for loop look at current and last period
        previous_dict = list_of_dict[i-1]
        current_dict = list_of_dict[i]
        #generate new variables to reference the value for calculation
        previous_amount = previous_dict["amount"]
        current_amount = current_dict["amount"]
        #calculate difference and change to integers
        difference = int(current_amount) - int(previous_amount)
        #create new dictionary referencing the difference in P/L and the current month's date
        dict_of_change = {
            "change" : difference,
            "date" : current_dict["date"],
        }
        #append new dictionaries to a new list that houses change between periods and current period
        list_of_change.append(dict_of_change)

#CALCULATE MIN, MAX, AND AVERAGE
#-------------------------------------------------------
    #calculate average change in P&L by getting changes from the list of change
    changes = [value['change'] for value in list_of_change] 
    #calculate average and round to 2 decimals and then calculate max and calculate min
    average = round(sum(changes) / len(changes),2)
    maximum = max(changes)
    minimum = min(changes)

#FIND DATES FOR MIN AND MAX BASED ON REFERENCING A THE DICTIONARY
#-------------------------------------------------------
    #determine the date for maximum change period in maximum calculation
    for value in list_of_change:
        #set the values within the dictionary to reference
        if 'change' in value and 'date' in value:
            change = value['change']
            date = value['date']
            #if the change is larger than the current maximum, set as new maximum and set the date as the new max date, otherwise ignore
            if change > maximum_change:
                maximum_change = change
                maximum_change_date = date
            #if the change is smaller than the current minimum, set as new minimum and set the date as the new min date, otherwise ignore
            if change < minimum_change:
                minimum_change = change
                minimum_change_date = date

#QUALITY CHECK, SAVING HERE
#-------------------------------------------------------
#check numbers & lists here through process             
#print(maximum)
#print(maximum_change_date)
#print(minimum)
#print(minimum_change_date)
#print(list_of_change)

#PRINT SUMMARY FOR ASSIGNMENT
#-------------------------------------------------------
#print summary
print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {total_months}")
print(f"Total Profit & Loss: ${total_p_and_l}")
print(f"Average  Change: ${average}")
print(f"Greatest Increase in Profits: {maximum_change_date} (${maximum})")
print(f"Greatest Decrease in Profits: {minimum_change_date} (${minimum})")

#EXPORT TO A FILE
#-------------------------------------------------------
#print to a text file and save, add a new line between each print
with open('FinalResults.txt', 'w') as f:
    f.write("Financial Analysis"'\n')
    f.write("----------------------------"'\n')
    f.write(f"Total Months: {total_months}"'\n')
    f.write(f"Total Profit & Loss: ${total_p_and_l}"'\n')
    f.write(f"Average  Change: ${average}"'\n')
    f.write(f"Greatest Increase in Profits: {maximum_change_date} (${maximum})"'\n')
    f.write(f"Greatest Decrease in Profits: {minimum_change_date} (${minimum})"'\n')
#close file
f.close()
```
